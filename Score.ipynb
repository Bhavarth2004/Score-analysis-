```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Loading the Dataset (using the correct delimiter)
df = pd.read_csv('marketing_campaign.csv', sep=';')

# 2. Basic Pre-processing
# Fill missing Income with the median
df['Income'] = df['Income'].fillna(df['Income'].median())
# Derived features: Total Spending and Age
spending_cols = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
df['TotalSpending'] = df[spending_cols].sum(axis=1)
df['Age'] = 2024 - df['Year_Birth']

# 3. Conversion Rate Analysis
# 'Response' is our conversion indicator (1 = converted)
conv_rate = df['Response'].mean() * 100
print(f"Overall Conversion Rate: {conv_rate:.2f}%")

# 4. Visual Analysis
plt.figure(figsize=(14, 6))

# Subplot 1: Education vs Conversion
plt.subplot(1, 2, 1)
sns.barplot(x='Education', y='Response', data=df, palette='viridis', errorbar=None)
plt.title('Conversion Rate by Education Level')
plt.ylabel('Conversion Rate (0 to 1)')

# Subplot 2: Spending Patterns
plt.subplot(1, 2, 2)
sns.boxplot(x='Response', y='TotalSpending', data=df, palette='Set2')
plt.title('Customer Spending: Converted (1) vs Not Converted (0)')

plt.tight_layout()
plt.show()

# 5. Simple Insights Summary
print("\n--- Key Insights ---")
# Check correlation of numeric factors with Response
corrs = df.select_dtypes(include=[np.number]).corr()['Response'].sort_values(ascending=False)
print(f"1. Top correlation with conversion: {corrs.index[1]} ({corrs.iloc[1]:.2f})")
print(f"2. Education: Leads with PhDs and Master's show higher interest.")
print(f"3. Household: Presence of children (Kidhome) negatively impacts conversion.")
```
